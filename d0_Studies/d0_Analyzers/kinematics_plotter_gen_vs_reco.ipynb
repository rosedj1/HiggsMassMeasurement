## This code generates many PDFs. Each PDF has multiple pages. On each page of a given PDF, you'll find a histogram of gen_lepX kinematics and reco_lepX kinematics (X could either be 1 or 2). A ratio plot of (reco-gen)/gen is at the bottom of each page. The entire PDF has the same cuts on $p_T$ and $\lvert \eta \rvert$, but each page differs in $d_{0}\times \mathrm{charge}$ cuts.
